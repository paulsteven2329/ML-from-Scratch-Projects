```python
import pandas as pd

from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier

from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline

from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
```

```python
data = {
    "ticket_text": [
        "My payment failed and money was deducted urgent help",
        "Cannot login to my account please help",
        "App is slow sometimes",
        "Need invoice copy for last month",
        "Server down for my business account urgent",
        "How to change my profile picture",
        "Refund not received for cancelled order",
        "Website is not loading urgent issue",
        "I want to update my phone number",
        "Account hacked please block immediately",
        "Feature request for dashboard",
        "Password reset link not working",
        "VIP customer unable to access product",
        "Minor spelling mistake in invoice",
        "Payment gateway error for all users urgent",
        "Need help understanding pricing plan",
        "Customer data missing after update",
        "Slow response from support team",
        "Business account billing problem urgent",
        "General question about subscription"
    ],
    "customer_type": [
        "Premium", "Free", "Free", "Free", "Enterprise",
        "Free", "Premium", "Enterprise", "Free", "Premium",
        "Free", "Free", "Enterprise", "Free", "Enterprise",
        "Free", "Premium", "Free", "Enterprise", "Free"
    ],
    "hours_waiting": [
        2, 5, 10, 12, 1,
        20, 6, 1, 15, 1,
        24, 8, 2, 30, 1,
        18, 3, 16, 2, 22
    ],
    "previous_tickets": [
        3, 1, 0, 0, 5,
        0, 2, 4, 1, 3,
        0, 1, 6, 0, 7,
        0, 4, 1, 5, 0
    ],
    "priority": [
        "High", "Medium", "Low", "Low", "High",
        "Low", "Medium", "High", "Low", "High",
        "Low", "Medium", "High", "Low", "High",
        "Low", "High", "Low", "High", "Low"
    ]
}

df = pd.DataFrame(data)
df.head()
```

```python
X = df.drop("priority",axis=1)
y = df["priority"]
```

```python
X_train, X_test, y_train , y_test = train_test_split(X,y,test_size=0.25,random_state=42)
```

```python
preprocessor = ColumnTransformer(
    transformers=[
        ("text",TfidfVectorizer(),"ticket_text"),
        ("customer",OneHotEncoder(handle_unknown="ignore"),["customer_type"])
    ],
    remainder="passthrough"
)
```

```python
model = Pipeline(
    steps=[
        ("preprocessor",preprocessor),
        ("classifer",RandomForestClassifier(
            n_estimators=100,
            random_state=42
        ))
    ]
)
```

```python
model.fit(X_train,y_train)
```

```python
y_pred = model.predict(X_test)

print("Accuracy:", accuracy_score(y_test,y_pred))
print()
print(classification_report(y_test,y_pred))
```

```python
confusion_matrix(y_test,y_pred)
```

```python
new_ticket = pd.DataFrame({
    "ticket_text": ["Payment failed and customer cannot access service urgent"],
    "customer_type": ["Enterprise"],
    "hours_waiting": [1],
    "previous_tickets": [5]
})

prediction = model.predict(new_ticket)
print("Predicted Priority:",prediction[0])
```
